```python
img_dict = {}
images = []
for f in files:
    out = ""
    out += "const PROGMEM uint32_t  " + os.path.basename(os.path.normpath(f.split(".")[0])) + " [2048]= "
    print("const PROGMEM uint32_t  " + os.path.basename(os.path.normpath(f.split(".")[0])) + " [2048]= ", end="")
    
    blank_image = np.zeros((32,64,3), np.uint8)
    image = cv2.imread(f)
    height, width, _ = image.shape
    resized_width = int(32*width/height)
    resized = cv2.resize(image, (resized_width, 32))
    x_offset = int((64-resized_width)/2)
    y_offset = 0
    blank_image[y_offset:x_offset+resized.shape[0], x_offset:x_offset+resized_width] = resized
    cv2.imwrite(f+".png", blank_image)
    cur = cv2.imread(f+".png")  
    out += "{"
    print("{", end="")
    tot = 0
    out += "\n"
    out += "\t"
    for i in range(0, 32):
        for j in range(0, 64):
            
#             print(tot, end=" :")
#             print(int(cur[i, j][0]))
            numbers = '0x%02x%02x%02x' % (int(cur[i, j][2]), int(cur[i, j][1]), int(cur[i, j][0])) + ","
            tot += 1
            out += numbers
            print(numbers, end="")
        print()
        out += "\n"
        out += "\t"
        
    out += "}"
    print("};")
    cp.copy(out)
    print(tot)
#     __ = input('press enter')
```
